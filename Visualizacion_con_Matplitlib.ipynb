# Visualizaciones Utiles con Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('/content/diabetes_train.csv')
df.head()
```

ahora hacemos un heatmap de correlaci√≥n entre variables con **seaborn**:

```python
plt.figure(figsize=(12,10), dpi= 80)
sns.heatmap(df.corr(), xticklabels=df.corr().columns, yticklabels=df.corr().columns, cmap='RdYlGn', center=0, annot=True)

# Decorations
plt.title('Correlograma de Diabetes', fontsize=22)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

An√°lisis visual de correlaci√≥n por combinaci√≥n de variables, separadas por clase (color), CON regresi√≥n:

```python
plt.figure(figsize=(10,8), dpi= 80)
sns.pairplot(df, kind="reg", hue="Diabetes")
plt.show()
```

```python
plt.figure(figsize=(10,8), dpi= 80)
sns.pairplot(df, kind="scatter", hue="Diabetes", plot_kws=dict(s=80, edgecolor="white", linewidth=2.5))
plt.show()
```
